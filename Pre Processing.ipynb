# 5. Check and handle outliers in at least 3 columns in the dataset

```python
sns.boxplot(data=data[['Age', 'Fare', 'SibSp']])
plt.show()
```

```python
data_with_outliers = data.copy(deep=True)
```

```python
def outlier_handling(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    q_limit = 1.5 * iqr
    upper_limit = q3 + q_limit
    lower_limt = q1 - q_limit
    outliers = data.loc[(data < lower_limt) ^ (data > upper_limit)].tolist()
    idx = data.loc[(data < lower_limt) ^ (data > upper_limit)].index.tolist()
    return outliers, idx
```

```python
fare_outlier_list, fare_outlier_index = outlier_handling(data['Fare'])
age_outlier_list, age_outlier_index = outlier_handling(data['Age'])
sibsp_oulliers, sibsp_outlier_index = outlier_handling(data['SibSp'])
```

```python
print(len(fare_outlier_index), len(age_outlier_index), len(sibsp_outlier_index))
```

```python
new_idx = sibsp_outlier_index + age_outlier_index + fare_outlier_index
len(new_idx)
```

```python
new_idx = set(new_idx)
len(new_idx)
```

```python
index_to_remove = list(new_idx)
data.drop(index=index_to_remove, inplace=True)
```
